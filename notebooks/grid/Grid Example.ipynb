```python
highDropout = nn.Sequential([
    nn.Linear(3,4),
    nn.ReLU(),
    nn.Dropout(0.10),
    nn.Linear(4,2),
    nn.Softmax(1),
    nn.Log()
])

lowDropout = nn.Sequential([
    nn.Linear(3,4),
    nn.ReLU(),
    nn.Dropout(0.05),
    nn.Linear(4,2),
    nn.Softmax(1),
    nn.Log()
])

g = grid.Grid()

config1 = g.configuration(
  model=highDropout,
  lr=0.01
)

config2 = g.configuration(
  model=lowDropout,
  lr=1
)

config3 = g.configuration(
  model=lowDropout,
  lr=0.1
)
```

```python
input = "IPFS address 1"
target = "IPFS address 2"

g.learn(
  input = input,
  target = target,
  configurations = [config1, config2, config3]
)
```
